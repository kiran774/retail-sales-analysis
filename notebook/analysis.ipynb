```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

# Load Dataset 
df = pd.read_csv("../data/superstore/Superstore.csv", encoding="latin1")
df.head()
   # Step 2  DATA CLEANING
#2a: Check for missing values
print("Missing values per column:/n", df.isnull().sum())

#2b: Remove duplicate rows 
df.drop_duplicates(inplace=True)
print("\nDuplicates removed. Total rows now:", len(df))

#2c: Check data types before conversion
print("\nData Types Before conversion:\n", df.dtypes)

#Convert 'order date to time safely
# First detect the column name
if 'Order Date' in df.columns:
   df.rename(columns={'Order Date': 'Order_Date'}, inplace=True)
elif 'order_date' in df.columns:
   df.rename(columns={'order_date': 'Order_Date'}, inplace=True)

df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
# Check data types after conversions
print("\nData types after conversion:\n", df.dtypes)

       # STEP 4   ANALYSIS
# Top performing Region by Total by total profit
top_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
print("Top performing region by profit:\n", top_region)

# Loss Making Product count
loss_products = df[df['Profit'] < 0]['Product Name'].value_counts()
print("\nLoss making Produts:\n", loss_products)

# Monthly sales trend (Total salesper month)
df['Month'] = df['Order_Date'].dt.to_period('M')  # Convert to year month period
monthly_sales = df.groupby('Month')['Sales'].sum()
print("\nMonthly Sales trend:\n", monthly_sales)
    
       # Step 5    VISULIZATION 
# Region vs profit (Bar Chart)
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=region_profit.index, y=region_profit.values, palette='viridis')
plt.title('Region vs Profit')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.grid(axis='y')
plt.show()

# Monthly Sales  (Line Chart)
df['Month'] = df['Order_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(12,5))
monthly_sales.plot(marker='o', linestyle='-', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Category contribution (Pie Chart)
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(7,7))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.show()


        #Step 6    INSIGHTs
#  Sales VS Profit relationship (Scatter Plot)
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title('Sales Vs Profit RelationShip')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()

# Sub_category Performance (Profit by Sub Category)
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=sub_profit.values, y=sub_profit.index)
plt.title('Profit by Sub-Category')
plt.xlabel('Total Profit')
plt.ylabel('Sub-Category')
plt.grid(axis='x')
plt.show()
```

## INSIGHTS & RECOMMENDATIONS  
1. **Most Profitable Region**
West Region makes the highest profit. Focus more on marketing and inventory in this region.

2. **Category To Improve**
Furniture category earn less profit. Improve pricing and cost management.

3. **Effects Of High Discounts**
Products with high discounts often give low or negative profit. Apply discounts carefully to keep profit stable.

4.  **Top Performing Produts**
Some Products, like Chair and Phones, give consistent high profit. Maintain and Promotion for these Produts.

5.  **Seasonal Trend**
Sales Peak in November-December. Plan inventory mad marketting according to seasonal demand.
